```python
pd.merge(df,inst_df,how="outer",on="Name")
```

```python
pd.merge(df,inst_df,how="left",on="Name")
```

```python
pd.merge(df,inst_df,how="right",on="Name")
```

```python
inst2 = inst_df.copy()
inst2 = inst2.rename(columns={'Name':'first'})
```
